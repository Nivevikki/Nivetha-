```python
import pandas as pd
import mysql.connector  # Uncomment this when using MySQL

# List of CSV file paths
csv_files = [
    "wb2_bus_details.csv", "Telangana_bus_details.csv", "rajasthan_bus_details.csv", 
    "ap_bus_details.csv", "assam_bus_details.csv", "chandigarh_bus_details.csv",  
    "himachal_bus_details.csv", "kerala_bus_details.csv", "kadamba_bus_details.csv", 
    "up_bus_details.csv"
]

# Read each CSV file into a DataFrame and store them in a list
dfs = [pd.read_csv(file) for file in csv_files]

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Add an 'id' column to the concatenated DataFrame
combined_df['id'] = range(1, len(combined_df) + 1)

# Save the concatenated DataFrame to a new CSV file
combined_df.to_csv("bus_routes.csv", index=False)

# Load the new combined DataFrame
df = pd.read_csv("bus_routes.csv")

# Replace 'INR ' in the 'Price' column with an empty string
df['Price'] = df['Price'].str.replace('INR ', '')

# Extract digits (0-9) from the 'Seat_Availability' column
df['Seat_Availability'] = df['Seat_Availability'].str.extract('(\d+)')

# Drop rows with any missing values (NaN)
df = df.dropna()

# Python MySQL connection using mysql.connector
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="redbus"
)

# Map pandas dtypes to SQL data types
dtype_mapping = {
    'float64': 'FLOAT',
    'object': 'TEXT',
    'int64': 'INT'
}

# Dynamically create the column definition for the SQL table
columns = ", ".join(f"{col} {dtype_mapping[str(dtype)]}" for col, dtype in zip(df.columns, df.dtypes))

# Define the table name as 'redbusproject'
table_name = "redbusproject"

# Create the table if it does not exist
with mydb.cursor() as cursor:
    cursor.execute(f"CREATE TABLE IF NOT EXISTS {table_name} ({columns})")

# Insert data from DataFrame into the MySQL table
insert_query = f"INSERT INTO {table_name} ({', '.join(df.columns)}) VALUES ({', '.join(['%s'] * len(df.columns))})"
data = [tuple(row) for row in df.itertuples(index=False)]

with mydb.cursor() as cursor:
    cursor.executemany(insert_query, data)
    mydb.commit()

# Close the database connection
mydb.close()

# Display the final DataFrame to verify
print(df)
```
